```python
# Aplicação do algoritmo K-means com 4 clusters
kmeans = KMeans(n_clusters=4)
kmeans.fit(X_scaled)
```

```python
# Aplicação do algoritmo K-means com diferentes números de clusters
# Alterando o número de clusters para 4 e verificar o que acontece com os grupos formados
kmeans_4 = KMeans(n_clusters=4)
kmeans_4.fit(X_scaled)
```

```python
# Alterando o número de clusters para 8 e verificar o que acontece com os grupos formados
kmeans_8 = KMeans(n_clusters=8)
kmeans_8.fit(X_scaled)
```
